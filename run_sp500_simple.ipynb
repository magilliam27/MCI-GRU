```python
# Verify uploads
import os
print("Files:")
for f in os.listdir('.'):
    if f.endswith('.py') or f.endswith('.csv'):
        print(f"  {f}")
```

```python
# Create a clean version of sp500.py that uses the uploaded CSV
with open('sp500.py', 'r') as f:
    lines = f.readlines()

new_lines = []
skip_until_filename = False

for line in lines:
    # Skip the download section
    if 'Loading tickers and downloading' in line:
        skip_until_filename = True
        new_lines.append('# Download section skipped - using uploaded CSV\n')
        new_lines.append('filename = "sp500_yf_download.csv"\n')
        continue
    if skip_until_filename:
        if 'Saved downloaded data' in line:
            skip_until_filename = False
        continue
    new_lines.append(line)

with open('sp500_run.py', 'w') as f:
    f.writelines(new_lines)

print('Created sp500_run.py (download section removed)')
```

```python
# Run the model
%run sp500_run.py
```
