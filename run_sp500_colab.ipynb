```python
# Patch sp500.py to skip download and use existing CSV
import re

with open('sp500.py', 'r') as f:
    code = f.read()

# Comment out the download section
code = re.sub(
    r"print\(.*Loading tickers and downloading.*\n.*load_tickers_from_csv.*\n.*download_sp500_dataframe.*\n.*\.to_csv.*\n.*print.*Saved downloaded.*",
    '# Download skipped - using pre-uploaded CSV\nfilename = "sp500_yf_download.csv"',
    code,
    flags=re.DOTALL
)

with open('sp500_patched.py', 'w') as f:
    f.write(code)

print('Patched sp500.py -> sp500_patched.py')
```
