## 4. Define Training Factor Matrix

Primary matrix: `5 objective settings x 2 label representations x 2 regime modes = 20 full-budget runs`.

Controls add edge-dropout and checkpoint-selection probes around the baseline objective, without crossing every lever into an oversized factorial.


```python
import itertools
import json

STATIC_MOMENTUM_OVERRIDES = [
    'features.include_momentum=true',
    'features.include_weekly_momentum=true',
    'features.momentum_encoding=binary',
    'features.momentum_blend_mode=static',
    'features.momentum_blend_fast_weight=0.5',
]

STATIC_THRESHOLD_GRAPH_OVERRIDES = [
    'graph.judge_value=0.8',
    'graph.update_frequency_months=0',
    'graph.corr_lookback_days=252',
    'graph.top_k=0',
    'graph.top_k_metric=corr',
    'graph.use_multi_feature_edges=true',
    'graph.append_snapshot_age_days=false',
    'graph.use_lead_lag_features=false',
    'graph.drop_edge_p=0.1',
    'training.shuffle_train=true',
]

def regime_overrides(enabled: bool, include_subsequent_returns: bool) -> list[str]:
    overrides = [
        f'features.include_global_regime={str(enabled).lower()}',
        f'features.regime_strict={str(REGIME_STRICT_FOR_REGIME_RUNS and enabled).lower()}',
    ]
    if enabled:
        overrides.extend([
            f'features.regime_include_subsequent_returns={str(include_subsequent_returns).lower()}',
            'features.regime_subsequent_return_horizons=[1,3]',
            'features.regime_change_months=12',
            'features.regime_norm_months=120',
            'features.regime_exclusion_months=1',
            'features.regime_similarity_quantile=0.2',
            'features.regime_min_history_months=24',
            f'features.regime_enforce_lag_days={REGIME_ENFORCE_LAG_DAYS}',
        ])
        if REGIME_INPUTS_CSV:
            overrides.append(f'features.regime_inputs_csv={REGIME_INPUTS_CSV}')
    return overrides

REGIME_FACTORS = [
    {
        'factor': 'regime_current_only',
        'description': 'Global regime similarity variables without historical subsequent-return features.',
        'overrides': regime_overrides(True, False),
    },
    {
        'factor': 'regime_with_forward_context',
        'description': 'Global regime similarity variables plus similar-history subsequent-return context.',
        'overrides': regime_overrides(True, True),
    },
]

OBJECTIVE_FACTORS = [
    {
        'factor': 'combined_alpha_025',
        'loss_type': 'combined',
        'ic_loss_alpha': 0.25,
        'description': 'Combined MSE + IC loss with light IC weighting.',
        'overrides': ['training.loss_type=combined', 'training.ic_loss_alpha=0.25'],
    },
    {
        'factor': 'combined_alpha_050',
        'loss_type': 'combined',
        'ic_loss_alpha': 0.50,
        'description': 'Current baseline combined MSE + IC loss.',
        'overrides': ['training.loss_type=combined', 'training.ic_loss_alpha=0.5'],
    },
    {
        'factor': 'combined_alpha_075',
        'loss_type': 'combined',
        'ic_loss_alpha': 0.75,
        'description': 'Combined loss with heavier IC weighting for rank alignment.',
        'overrides': ['training.loss_type=combined', 'training.ic_loss_alpha=0.75'],
    },
    {
        'factor': 'pure_ic',
        'loss_type': 'ic',
        'ic_loss_alpha': None,
        'description': 'Pure IC loss; high rank alignment, possibly less stable.',
        'overrides': ['training.loss_type=ic'],
    },
    {
        'factor': 'pure_mse',
        'loss_type': 'mse',
        'ic_loss_alpha': None,
        'description': 'Pure MSE loss; useful control for magnitude-target training.',
        'overrides': ['training.loss_type=mse'],
    },
]

LABEL_FACTORS = [
    {
        'factor': 'returns_label_5d',
        'label_type': 'returns',
        'label_t': 5,
        'description': 'Raw 5-day forward returns, matching current baseline.',
        'overrides': ['training.label_type=returns', 'model.label_t=5'],
    },
    {
        'factor': 'rank_label_5d',
        'label_type': 'rank',
        'label_t': 5,
        'description': 'Cross-sectional 5-day rank percentile labels.',
        'overrides': ['training.label_type=rank', 'model.label_t=5'],
    },
]

def clean_name(*parts):
    return '__'.join(part.replace('_', '-') for part in parts)

ABLATIONS = []
for objective, label, regime in itertools.product(OBJECTIVE_FACTORS, LABEL_FACTORS, REGIME_FACTORS):
    ABLATIONS.append({
        'name': clean_name(objective['factor'], label['factor'], regime['factor']),
        'description': ' | '.join([objective['description'], label['description'], regime['description']]),
        'family': 'primary',
        'objective_factor': objective['factor'],
        'label_factor': label['factor'],
        'regime_factor': regime['factor'],
        'loss_type': objective['loss_type'],
        'ic_loss_alpha': objective['ic_loss_alpha'],
        'label_type': label['label_type'],
        'label_t': label['label_t'],
        'selection_metric': 'val_ic',
        'drop_edge_p': 0.1,
        'overrides': [
            *STATIC_MOMENTUM_OVERRIDES,
            *STATIC_THRESHOLD_GRAPH_OVERRIDES,
            *objective['overrides'],
            *label['overrides'],
            *regime['overrides'],
            'training.selection_metric=val_ic',
        ],
    })

if INCLUDE_REGULARIZATION_CONTROLS:
    for drop_edge_p, regime in itertools.product([0.0, 0.05, 0.2], REGIME_FACTORS):
        ABLATIONS.append({
            'name': clean_name(f'drop_edge_{str(drop_edge_p).replace(".", "p")}', 'combined_alpha_050', 'returns_label_5d', regime['factor']),
            'description': f'Edge dropout control at p={drop_edge_p} around the current combined-loss baseline.',
            'family': 'regularization_control',
            'objective_factor': 'combined_alpha_050',
            'label_factor': 'returns_label_5d',
            'regime_factor': regime['factor'],
            'loss_type': 'combined',
            'ic_loss_alpha': 0.5,
            'label_type': 'returns',
            'label_t': 5,
            'selection_metric': 'val_ic',
            'drop_edge_p': drop_edge_p,
            'overrides': [
                *STATIC_MOMENTUM_OVERRIDES,
                *STATIC_THRESHOLD_GRAPH_OVERRIDES,
                'training.loss_type=combined',
                'training.ic_loss_alpha=0.5',
                'training.label_type=returns',
                'model.label_t=5',
                'training.selection_metric=val_ic',
                f'graph.drop_edge_p={drop_edge_p}',
                *regime['overrides'],
            ],
        })

if INCLUDE_SELECTION_CONTROLS:
    for objective_name, objective_overrides, regime in itertools.product(
        ['combined_alpha_050', 'combined_alpha_075'],
        [['training.loss_type=combined', 'training.ic_loss_alpha=0.5'], ['training.loss_type=combined', 'training.ic_loss_alpha=0.75']],
        REGIME_FACTORS,
    ):
        ABLATIONS.append({
            'name': clean_name('selection_val_loss', objective_name, 'returns_label_5d', regime['factor']),
            'description': 'Checkpoint-selection control using validation loss instead of validation IC.',
            'family': 'selection_control',
            'objective_factor': objective_name,
            'label_factor': 'returns_label_5d',
            'regime_factor': regime['factor'],
            'loss_type': 'combined',
            'ic_loss_alpha': 0.5 if objective_name.endswith('050') else 0.75,
            'label_type': 'returns',
            'label_t': 5,
            'selection_metric': 'val_loss',
            'drop_edge_p': 0.1,
            'overrides': [
                *STATIC_MOMENTUM_OVERRIDES,
                *STATIC_THRESHOLD_GRAPH_OVERRIDES,
                *objective_overrides,
                'training.label_type=returns',
                'model.label_t=5',
                'training.selection_metric=val_loss',
                *regime['overrides'],
            ],
        })

print('Primary runs:', sum(a['family'] == 'primary' for a in ABLATIONS))
print('Regularization controls:', sum(a['family'] == 'regularization_control' for a in ABLATIONS))
print('Selection controls:', sum(a['family'] == 'selection_control' for a in ABLATIONS))
print('Total queued runs before filters:', len(ABLATIONS))
pd.DataFrame(ABLATIONS)[[
    'name', 'family', 'objective_factor', 'label_factor', 'regime_factor',
    'loss_type', 'ic_loss_alpha', 'label_type', 'label_t', 'selection_metric', 'drop_edge_p', 'description'
]]
```

## 5. Optional Label Horizon Probes

The main matrix keeps `model.label_t=5` so it remains comparable to prior runs. Longer horizons need wider train/validation/test embargo gaps. Flip `INCLUDE_LABEL_HORIZON_PROBES=True` only if you are comfortable comparing those probes on the adjusted split dates below.


```python
if INCLUDE_LABEL_HORIZON_PROBES:
    HORIZON_VAL_START = '2024-02-01'
    HORIZON_TEST_START = '2025-02-01'
    horizon_objectives = [
        ('combined_alpha_050', ['training.loss_type=combined', 'training.ic_loss_alpha=0.5'], 0.5),
        ('combined_alpha_075', ['training.loss_type=combined', 'training.ic_loss_alpha=0.75'], 0.75),
    ]
    for label_t, label_type, objective, regime in itertools.product(
        [10, 20],
        ['returns', 'rank'],
        horizon_objectives,
        REGIME_FACTORS,
    ):
        objective_name, objective_overrides, alpha = objective
        ABLATIONS.append({
            'name': clean_name(f'horizon_{label_t}d', f'{label_type}_label', objective_name, regime['factor']),
            'description': f'{label_t}-day {label_type} label probe with adjusted embargo-safe split dates.',
            'family': 'horizon_probe',
            'objective_factor': objective_name,
            'label_factor': f'{label_type}_label_{label_t}d',
            'regime_factor': regime['factor'],
            'loss_type': 'combined',
            'ic_loss_alpha': alpha,
            'label_type': label_type,
            'label_t': label_t,
            'selection_metric': 'val_ic',
            'drop_edge_p': 0.1,
            'overrides': [
                *STATIC_MOMENTUM_OVERRIDES,
                *STATIC_THRESHOLD_GRAPH_OVERRIDES,
                *objective_overrides,
                f'training.label_type={label_type}',
                f'model.label_t={label_t}',
                f'data.val_start={HORIZON_VAL_START}',
                f'data.test_start={HORIZON_TEST_START}',
                'training.selection_metric=val_ic',
                *regime['overrides'],
            ],
        })
    print('Added horizon probes. New total:', len(ABLATIONS))
else:
    print('Horizon probes disabled; main matrix remains on 5-day labels.')
```

## 6. Helpers: Run, Collect, Score


```python
import numpy as np
from pathlib import Path

BASE_OVERRIDES = [
    'features=with_momentum',
    'data.source=csv',
    f'data.filename={DATA_FILE}',
    f'data.train_start={TRAIN_START}',
    f'data.train_end={TRAIN_END}',
    f'data.val_start={VAL_START}',
    f'data.val_end={VAL_END}',
    f'data.test_start={TEST_START}',
    f'data.test_end={TEST_END}',
    f'training.num_models={NUM_MODELS}',
    f'training.num_epochs={NUM_EPOCHS}',
    f'training.early_stopping_patience={EARLY_STOPPING_PATIENCE}',
    f'training.batch_size={BATCH_SIZE}',
    f'training.learning_rate={LEARNING_RATE}',
    f'evaluation.bootstrap_resamples={BOOTSTRAP_RESAMPLES}',
    'tracking.enabled=true',
    'tracking.log_predictions=false',
    f'seed={SEED}',
]

def run_ablation(ablation):
    run_dir = ABLATION_ROOT / ablation['name']
    run_dir.mkdir(parents=True, exist_ok=True)
    done_marker = run_dir / 'evaluation_summary.json'
    if SKIP_COMPLETED_RUNS and done_marker.exists():
        print('\n' + '=' * 100)
        print('Skipping completed run:', ablation['name'])
        print('Run dir:', run_dir)
        print('=' * 100)
        return {'name': ablation['name'], 'run_dir': str(run_dir), 'returncode': 0, 'skipped': True}

    cmd = [
        sys.executable,
        '-u',
        'run_experiment.py',
        *BASE_OVERRIDES,
        *ablation['overrides'],
        f'experiment_name={ablation["name"]}',
        f'hydra.run.dir={run_dir}',
    ]
    print('\n' + '=' * 100)
    print('Running:', ablation['name'])
    print(ablation['description'])
    print('Command:', ' '.join(cmd))
    print('=' * 100)
    result = subprocess.run(cmd, cwd=REPO_DIR, text=True, capture_output=True)
    (run_dir / 'stdout.log').write_text(result.stdout)
    (run_dir / 'stderr.log').write_text(result.stderr)
    print(result.stdout[-4000:])
    if result.returncode != 0:
        print(result.stderr[-4000:])
    return {'name': ablation['name'], 'run_dir': str(run_dir), 'returncode': result.returncode, 'skipped': False}

def flatten_json(prefix, obj, out):
    if isinstance(obj, dict):
        for key, value in obj.items():
            flatten_json(f'{prefix}.{key}' if prefix else key, value, out)
    elif isinstance(obj, list):
        out[prefix] = json.dumps(obj)
    else:
        out[prefix] = obj

def load_json(path):
    path = Path(path)
    if path.exists():
        return json.loads(path.read_text())
    return None

def collect_ablation(ablation, status):
    run_dir = Path(status['run_dir'])
    row = {
        'name': ablation['name'],
        'description': ablation['description'],
        'family': ablation.get('family'),
        'objective_factor': ablation.get('objective_factor'),
        'label_factor': ablation.get('label_factor'),
        'regime_factor': ablation.get('regime_factor'),
        'loss_type': ablation.get('loss_type'),
        'ic_loss_alpha': ablation.get('ic_loss_alpha'),
        'label_type': ablation.get('label_type'),
        'label_t': ablation.get('label_t'),
        'selection_metric': ablation.get('selection_metric'),
        'drop_edge_p': ablation.get('drop_edge_p'),
        'returncode': status['returncode'],
        'skipped': status.get('skipped', False),
        'run_dir': str(run_dir),
        'overrides': ' '.join(ablation['overrides']),
    }
    for filename, prefix in [
        ('training_summary.json', 'training'),
        ('evaluation_summary.json', 'evaluation'),
        ('run_metadata.json', 'metadata'),
        ('walkforward_summary.json', 'walkforward'),
    ]:
        data = load_json(run_dir / filename)
        if data is not None:
            flatten_json(prefix, data, row)
    return row

def add_decision_columns(df):
    out = df.copy()
    numeric_cols = [
        'evaluation.metrics.avg_ic',
        'evaluation.metrics.avg_ic_ci_lower',
        'evaluation.metrics.avg_spearman_corr',
        'evaluation.metrics.sharpe_top_20_newey_west',
        'evaluation.metrics.return_top_20',
        'evaluation.metrics.top_20_return_ci_lower',
        'evaluation.metrics.hit_rate',
        'evaluation.metrics.long_short_spread',
        'training.mean_best_val_ic',
        'training.models_trained',
        'ic_loss_alpha',
        'label_t',
        'drop_edge_p',
    ]
    for col in numeric_cols:
        if col in out.columns:
            out[col] = pd.to_numeric(out[col], errors='coerce')
    score = pd.Series(0.0, index=out.index)
    weights = {
        'evaluation.metrics.avg_ic': 0.35,
        'evaluation.metrics.avg_spearman_corr': 0.25,
        'evaluation.metrics.sharpe_top_20_newey_west': 0.25,
        'evaluation.metrics.return_top_20': 0.15,
    }
    for col, weight in weights.items():
        if col in out.columns:
            vals = out[col].astype(float)
            denom = vals.std(skipna=True)
            if denom and np.isfinite(denom) and denom > 0:
                score += weight * ((vals - vals.mean(skipna=True)) / denom).fillna(0.0)
    out['decision_score'] = score
    out['status'] = np.where(out['returncode'].eq(0), 'OK', 'FAILED')
    if 'evaluation.metrics.avg_ic_ci_lower' in out.columns:
        out['ic_ci_pass'] = out['evaluation.metrics.avg_ic_ci_lower'].gt(0)
    if 'evaluation.metrics.top_20_return_ci_lower' in out.columns:
        out['top20_ci_pass'] = out['evaluation.metrics.top_20_return_ci_lower'].gt(0)
    return out.sort_values(['status', 'decision_score'], ascending=[True, False])
```

## 7. Run Training Factor Matrix


```python
manifest_path = ABLATION_ROOT / 'training_factor_ablation_manifest.json'
manifest_path.write_text(json.dumps({
    'ablations': ABLATIONS,
    'base_overrides': BASE_OVERRIDES,
    'filters': {
        'run_only_families': RUN_ONLY_FAMILIES,
        'run_name_contains': RUN_NAME_CONTAINS,
        'skip_completed_runs': SKIP_COMPLETED_RUNS,
    },
    'factor_counts': {
        'objectives': len(OBJECTIVE_FACTORS),
        'labels': len(LABEL_FACTORS),
        'regimes': len(REGIME_FACTORS),
        'primary_runs': sum(a['family'] == 'primary' for a in ABLATIONS),
        'regularization_controls': sum(a['family'] == 'regularization_control' for a in ABLATIONS),
        'selection_controls': sum(a['family'] == 'selection_control' for a in ABLATIONS),
        'horizon_probes': sum(a['family'] == 'horizon_probe' for a in ABLATIONS),
    },
    'budget': {
        'num_models': NUM_MODELS,
        'num_epochs': NUM_EPOCHS,
        'early_stopping_patience': EARLY_STOPPING_PATIENCE,
        'bootstrap_resamples': BOOTSTRAP_RESAMPLES,
    },
}, indent=2))

run_queue = ABLATIONS
if RUN_ONLY_FAMILIES:
    run_queue = [a for a in run_queue if a.get('family') in set(RUN_ONLY_FAMILIES)]
if RUN_NAME_CONTAINS:
    run_queue = [a for a in run_queue if RUN_NAME_CONTAINS in a['name']]

print('Runs after filters:', len(run_queue))

statuses = []
rows = []
for ablation in run_queue:
    status = run_ablation(ablation)
    statuses.append(status)
    rows.append(collect_ablation(ablation, status))
    interim_df = add_decision_columns(pd.DataFrame(rows))
    interim_df.to_csv(ABLATION_ROOT / 'ablation_decision_table_interim.csv', index=False)

raw_df = pd.DataFrame(rows)
raw_path = ABLATION_ROOT / 'ablation_results_raw.csv'
raw_df.to_csv(raw_path, index=False)

decision_df = add_decision_columns(raw_df)
decision_path = ABLATION_ROOT / 'ablation_decision_table.csv'
decision_df.to_csv(decision_path, index=False)

html_path = ABLATION_ROOT / 'ablation_decision_table.html'
decision_df.to_html(html_path, index=False)

print('Saved manifest:', manifest_path)
print('Saved raw results:', raw_path)
print('Saved decision table:', decision_path)
print('Saved HTML table:', html_path)

display_cols = [c for c in [
    'status',
    'name',
    'family',
    'decision_score',
    'evaluation.metrics.avg_ic',
    'evaluation.metrics.avg_ic_ci_lower',
    'evaluation.metrics.avg_spearman_corr',
    'evaluation.metrics.sharpe_top_20_newey_west',
    'evaluation.metrics.return_top_20',
    'evaluation.metrics.top_20_return_ci_lower',
    'training.mean_best_val_ic',
    'run_dir',
] if c in decision_df.columns]
display(decision_df[display_cols])
```

## 8. Main Effects and Interaction Tables


```python
METRICS_FOR_EFFECTS = [
    'decision_score',
    'evaluation.metrics.avg_ic',
    'evaluation.metrics.avg_ic_ci_lower',
    'evaluation.metrics.avg_spearman_corr',
    'evaluation.metrics.return_top_20',
    'evaluation.metrics.top_20_return_ci_lower',
    'evaluation.metrics.sharpe_top_20_newey_west',
]
METRICS_FOR_EFFECTS = [c for c in METRICS_FOR_EFFECTS if c in decision_df.columns]

ok_runs = decision_df[decision_df['status'].eq('OK')].copy()
ok_primary = ok_runs[ok_runs['family'].eq('primary')].copy()

effect_tables = {}
for factor in [
    'objective_factor',
    'loss_type',
    'ic_loss_alpha',
    'label_factor',
    'label_type',
    'regime_factor',
    'selection_metric',
    'drop_edge_p',
    'family',
]:
    source = ok_primary if factor in {'objective_factor', 'loss_type', 'ic_loss_alpha', 'label_factor', 'label_type', 'regime_factor'} else ok_runs
    if factor in source.columns and not source.empty:
        effect = source.groupby(factor, dropna=False)[METRICS_FOR_EFFECTS].agg(['mean', 'median', 'count'])
        effect_tables[factor] = effect
        path = ABLATION_ROOT / f'{factor}_main_effects.csv'
        effect.to_csv(path)
        print('Saved:', path)
        display(effect)

if not ok_primary.empty and METRICS_FOR_EFFECTS:
    interaction_pairs = [
        ('objective_factor', 'label_factor'),
        ('objective_factor', 'regime_factor'),
        ('label_factor', 'regime_factor'),
        ('loss_type', 'label_type'),
    ]
    for left, right in interaction_pairs:
        if left in ok_primary.columns and right in ok_primary.columns:
            pivot = ok_primary.pivot_table(
                index=left,
                columns=right,
                values='decision_score',
                aggfunc='mean',
            )
            path = ABLATION_ROOT / f'interaction__{left}__{right}.csv'
            pivot.to_csv(path)
            print('Saved:', path)
            display(pivot)
else:
    print('No successful primary runs available for effect tables yet.')
```
